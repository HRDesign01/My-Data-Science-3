```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

HEADER TAGS FROM https://en.wikipedia.org/wiki/Main_Page

```python
Header_titles = soup.find('span',class_="mw-headline")
Header_titles
```

```python
Header_titles.text
```

```python
Headers = []

for i in soup.find_all('span', class_="mw-headline"):
    Headers.append(i.text)
    
Headers
```

```python
import pandas as pd
df = pd.DataFrame({'Header':Headers})
df
```
